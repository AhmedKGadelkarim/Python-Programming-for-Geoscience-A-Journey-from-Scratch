```python
phi = [0.2, 0.3, 0.25, 0.4, 0.2, 0.35, 0.31, 0.30, 0.2]
correction = [0.02, 0.05, 0.03, 0.03, 0.07, 0.05, 0.02, 0.05, 0.01]
i = 0
while i <len(phi):
    corrected_porosity = (phi[i] - correction[i])
    print(f"{corrected_porosity:.3f}")
    i +=1
```

```python
for i in range(len(phi)):
    corrected_porosity = (phi[i] - correction[i])
    print(f"{corrected_porosity:.3f}")
```

```python
corrected_porosity
```

```python
corrected_porosity_list = []
for i in range(len(phi)):
    corrected_porosity = (phi[i] - correction[i])
    corrected_porosity_list.append(round(corrected_porosity, 2))
    print(f"{corrected_porosity:.3f}")
```

```python
time = [30, 510 , 750, 1150, 1230, 1548, 1750, 1790, 1853, 1910, 2060, 2100]  #milliseconds

velocities = [1450, 1700, 1720, 1900, 1940, 2000, 2100, 2200, 2400, 3000, 2300, 2600]   #m/s
```

```python
def time_to_depth(TWT):
    depth = []
    for i in range(len(TWT)):
        depth.append(TWT[i] * velocities[i] / 2000)
    return depth
```

```python
time_to_depth(time)
```

```python
depth
```
