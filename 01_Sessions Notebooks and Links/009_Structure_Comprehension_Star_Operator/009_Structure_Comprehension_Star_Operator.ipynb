```python
wavelet = [0.2, 0.5, 0.8, 1.0, 0.7]
reflection_coeff = [0.1, 0.3, 0.6, 0.9, 0.5]
noise = [0.05, -0.02, 0.01, -0.03, 0.02]
```

```python
Seismic_Response = []

for i in range(len(wavelet)):
    Seismic_value = reflection_coeff[i] * wavelet[i] + noise[i]
    Seismic_Response.append(Seismic_value)

print(Seismic_Response)
```

```python
Seismic_Response = []
def Seismic_Response_Calc(reflection_coeff, wavelet, noise):
    for i in range(len(wavelet)):
        Seismic_value = reflection_coeff[i] * wavelet[i] + noise[i]
        Seismic_Response.append(Seismic_value)
    return Seismic_Response

Seismic_Response_Calc(reflection_coeff, wavelet, noise)
```

```python
Pos_amp = []

for i in Seismic_Response:
    if i >=0:
        Pos_amp.append(i)
    else:
        pass

print(Pos_amp)
```

```python
#Positive_amplitude
Positive_amplitude = [ amplitude for amplitude in Seismic_Response if amplitude>=0]

print(Positive_amplitude)
```

```python
#Absolute Amplitude

Abs_amp = []

for i in Seismic_Response:
    if i >=0:
        Abs_amp.append(i)
    else:
        Abs_amp.append(i*-1)

print(Abs_amp)
```

```python
Seismic_Response = [0.15, -0.17, 0.47, 0.93, -0.37, 0.28, -0.23, 0.57, -0.07, 0.13, 0.28, 0.26, 0.54, -0.7, -0.07, 0.36, -0.32, 0.44, 0.26, -0.19]
```

```python
high_amplitude = [ ]
low_amp = []


for amp in Seismic_Response:
    if amp >0.28 or amp <-0.28:
        high_amplitude.append(amp)
    elif -0.28 < amp <= 0.28:
        low_amp.append(amp)

print(high_amplitude)
print(low_amp)
```

```python
high_amplitude = [ amplitude for amplitude in Seismic_Response if amplitude > 0.28 or amplitude < -0.28]
high_amplitude
```

```python
high_amp_dict = {}
low_amp_dict = {}

for amp in Seismic_Response:
    if amp >0.28 or amp <-0.28:
        high_amp_dict[amp] = "High amplitude"
    elif -0.28 < amp <= 0.28:
        low_amp_dict[amp] = "Low Amplitude"
```
