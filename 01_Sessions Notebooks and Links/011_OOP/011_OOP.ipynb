```python
class Reservoir:
    formation_volume_factor = 1.3  # Common formation factor for all reservoirs

    def __init__(self, name, area, thickness, porosity, oil_saturation):
        print(f"{name} Reservoir object created!")  
        self.name = name
        self.area = area  # m²
        self.thickness = thickness  # m
        self.porosity = porosity  # fraction
        self.oil_saturation = oil_saturation  # fraction

    def calc_OOIP(self):
        # OOIP calculation (in cubic meters)
        OOIP = ((self.area * self.thickness * self.porosity * self.oil_saturation)/self.formation_volume_factor)
        #print(f"OOIP for {self.name} reservoir is {OOIP} m3")
        return(f"OOIP for {self.name} reservoir is {OOIP:.3f} m3")
```

```python
Reservoir_1 = Reservoir("Nukhul", 2000, 25, 0.18, 0.85)
```

```python
Reservoir_2 = Reservoir("Nubia", 3500, 20, 0.2, 0.9)
```

```python
Reservoir_1.calc_OOIP()
```

```python
Reservoir_2.calc_OOIP()
```

```python
#2_Encapsulation  : 
# 3 _ Name  Mangling  

class SeismicSurvey:
    def __init__(self, project_name, _survey_area, __budget):
        self.project_name = project_name  # Public attribute
        self._survey_area = _survey_area  # Protected attribute
        self.__budget = __budget  # Private attribute

    def display_project(self):
        """Public method to display project details."""
        print(f"Project Name: {self.project_name}")

    def _display_area(self):
        """Protected method to display the survey area (for internal use)."""
        print(f"Survey Area: {self._survey_area} sq km")

    def __display_budget(self):
        """Private method to display the project budget (not accessible externally)."""
        print(f"Project Budget: ${self.__budget}")

    def access_budget(self):
        """Public method to access the private budget."""
        self.__display_budget()
```

```python
class SeismicAttribute:
    def __init__(self, survey_name, depth, wave_velocity, attribute_name, attribute_value):
        self.survey = survey_name
        self.depth = depth
        self.velocity = wave_velocity
        self.attribute_name = attribute_name
        self.attribute_value  = attribute_value

    def describe_survey(self):
        return f"Survey: {self.survey}, Depth: {self.depth}m, Wave Velocity: {self.velocity} m/s"
    
    def describe_attribute(self):
        return f"{self.attribute_name} has a value of {self.attribute_value} for {self.survey} survey."
```

```python
class SeismicAttribute(SeismicData):
    def __init__(self, survey_name, depth, wave_velocity, attribute_name, attribute_value):
        super().__init__(survey_name, depth, wave_velocity)
        self.attribute_name = attribute_name
        self.attribute_value  = attribute_value
    def describe_attribute(self):
        return f"{self.attribute_name} has a value of {self.attribute_name} for {self.survey} survey."
```

```python
class AVO_attributes(SeismicAttribute):
    def __init__(self, survey_name, depth, wave_velocity, attribute_name, value, gradient, intercept):
        super().__init__(survey_name, depth, wave_velocity, attribute_name, value)
        self.gradient = gradient
        self.intercept = intercept

    def avo_result(self):
        return f"AVO Analysis: Gradient = {self.gradient}, Intercept = {self.intercept}"
```

```python
#Polymorphism : 


# Parent Class: SeismicData
class SeismicData:
    def __init__(self, survey_name, depth, wave_velocity):
        self.survey = survey_name
        self.depth = depth
        self.velocity = wave_velocity

    def describe_survey(self):
        return f"Survey: {self.survey}, Depth: {self.depth}m, Wave Velocity: {self.velocity} m/s"

# Child Class: SeismicAttribute (Extending SeismicData)
class SeismicAttribute(SeismicData):
    def __init__(self, survey_name, depth, wave_velocity, attribute_name, value):
        super().__init__(survey_name, depth, wave_velocity)
        self.attribute = attribute_name
        self.value = value

    # Polymorphism - Overriding describe_survey() to include attribute details
    def describe_survey(self):
        return (f"Survey: {self.survey}, Depth: {self.depth}m, Wave Velocity: {self.velocity} m/s, "
                f"Attribute: {self.attribute}, Value: {self.value}")

# Grandchild Class: AVOAnalysis (Extending SeismicAttribute)
class AVOAnalysis(SeismicAttribute):
    def __init__(self, survey_name, depth, wave_velocity, attribute_name, value, gradient, intercept):
        super().__init__(survey_name, depth, wave_velocity, attribute_name, value)
        self.gradient = gradient
        self.intercept = intercept

    # Further overriding describe_survey() to include AVO parameters
    def describe_survey(self):
        return (f"AVO Analysis for {self.survey}: Depth = {self.depth}m, Velocity = {self.velocity} m/s, "
                f"Attribute = {self.attribute}, Value = {self.value}, Gradient = {self.gradient}, "
                f"Intercept = {self.intercept}")

    def avo_result(self):
        return f"AVO Analysis: Gradient = {self.gradient}, Intercept = {self.intercept}"


```

```python
# Creating instances of each class
basic_survey = SeismicData("North Basin Survey", 2500, 3500)
attribute_survey = SeismicAttribute("North Basin Survey", 2500, 3500, "Acoustic Impedance", 12000)
avo_analysis = AVOAnalysis("North Basin Survey", 2500, 3500, "Amplitude", 0.85, -0.1, 0.3)
```
