```python
avo_class = []
def avo(a):
    index = 0
    if a[index+1] > a[index] and a[index+2] > a[index+1]:
        avo_class = list(("Class 3",))
    elif a[index+1] < a[index] and a[index+2] < a[index+1]:
        avo_class = list(("Class 1" , ))
    else:
        avo_class = list(("No AVO", ))
    return avo_class
```

```python
PG = [ [10, 20, 30], [20, 10, 5], [5, 10, 5], [5, 10, 15], [5, 5, 5]]

#List[sublist_Index][item_index]
```

```python
avo_class = []

def avo(a):
    index = 0
    sub_index = 0
    for index in range(len(a)):
        if a[index][sub_index+1] > a[index][sub_index] and a[index][sub_index+2] > a[index][sub_index+1]:
            avo_class.append('Class 3')
        elif a[index][sub_index+1] < a[index][sub_index] and a[index][sub_index+2] < a[index][sub_index+1]:
            avo_class.append('class 1')
        else:
            avo_class.append('No AVO')
    return avo_class
```
