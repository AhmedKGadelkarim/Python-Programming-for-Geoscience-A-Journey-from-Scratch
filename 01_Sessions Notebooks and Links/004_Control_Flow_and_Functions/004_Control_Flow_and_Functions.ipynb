```python
GR = float(input("What is GR Value ? "))

if GR < 0 or GR >200:
    print("Invalid Value")

elif GR > 45:
    print("Shale")
elif GR <= 45:
    Res = float(input("What is Resistivity Value ? "))
    if Res <2:
        print("Brine Sand")
    elif Res >=2:
        Density = float(input("What is Density Value ? "))
        Neutron = float(input("What is Neutron Value ? "))
        if Density <2.2 and Neutron <15:
            print("Gas Sand")
        elif Density >=2.2 and Neutron >=15:
            print("Oil Sand")
```

```python
#calculate total seismic travel time if seismic wave travelling though 4 layers

layers = int(input("Enter the number of layers"))
total_travel_time = 0

for layer in range(1, layers+1):
    thicknes = int(input(f'Enetr the thickness value for layer {layer}: '))
    velocity = int(input("Enter the velocity value for layer {layer}: "))
    travel_time = (2*thicknes)/velocity

    total_travel_time += travel_time
print(f"Total Travel time is {total_travel_time:.2f} seconds")
```

```python
#calculate total seismic travel time if seismic wave travelling though 4 layers

layers = int(input("Enter the number of layers"))
total_travel_time = 0
layer = 1

while layer <=layers:
    thicknes = int(input('Enetr the thickness value: '))
    velocity = int(input("Enter the velocity value: "))
    travel_time = (2*thicknes)/velocity

    total_travel_time += travel_time

    layer +=1
print(f"Total Travel time is {total_travel_time:.2f} seconds")
```
