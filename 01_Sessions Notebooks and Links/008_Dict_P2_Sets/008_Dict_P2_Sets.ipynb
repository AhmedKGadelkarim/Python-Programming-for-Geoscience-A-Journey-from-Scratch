```python
AVO_Class_3 = {'Bright spot', 'Flat spot', 'Amplitude shut off', 'Pull down'}

AVO_Class_1 = {'Dim spot', 'Pull down', 'Flat spot', }
```

```python
# Sets_Operations


#Union    >>> |   or    union()

Class_Union = AVO_Class_3 | AVO_Class_1

Class_Union
```

```python
Class_Union_2 = AVO_Class_3.union(AVO_Class_1)
print(Class_Union_2)
```

```python
#Set Intersection
# &      intersection()

Class_Intersection = AVO_Class_3 & AVO_Class_1
Class_Intersection

```

```python
Class_Inter = AVO_Class_3.intersection(AVO_Class_1)
Class_Inter
```

```python
#Set Difference


Dif_3_1 = AVO_Class_3 - AVO_Class_1

Dif_3_1
```

```python
Dif_1_3 = AVO_Class_1 - AVO_Class_3

Dif_1_3
```

```python
#Symmetric Differnce

# ^   or symmetric_difference()



Class_Sym_Diff = AVO_Class_3 ^ AVO_Class_1

Class_Sym_Diff
```

```python
Sym_Diff = AVO_Class_3.symmetric_difference(AVO_Class_1)

Sym_Diff
```
