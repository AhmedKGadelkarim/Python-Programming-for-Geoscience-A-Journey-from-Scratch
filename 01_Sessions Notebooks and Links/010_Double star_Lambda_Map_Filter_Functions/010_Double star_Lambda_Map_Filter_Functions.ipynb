```python
Intercept = [0.79, 0.561, 0.55, 0.277, -0.057, -0.122, 0.108, 0.103, 0.042, -0.269, -0.367, 0.661, 0.4, -0.78, 0.618, -0.079, -0.645, -0.484, 0.406, -0.373]

Gradient = [0.745, 0.629, -0.918, -0.567, 0.313, 0.52, 0.72, -0.407, 0.886, -0.202, 0.749, -0.601, 0.381, 0.639, 0.665, 0.859, -0.681, -0.317, -0.229, 0.063]

Theta_7 = 7
Theta_25 = 25
```

```python
import math
seismic_response_7 = []
seismic_response_25 = []

if Theta_7 == 7:
    for i in range(len(Intercept)):
        R = Intercept[i] + Gradient[i] * (math.sin(Theta_7) ** 2)
        seismic_response_7.append(round(R, 3))
if Theta_25 == 25:
    for i in range(len(Intercept)):
        R = Intercept[i] + Gradient[i] * (math.sin(Theta_25) ** 2)
        seismic_response_25.append(round(R, 3))

print("Seismic Response for Theta = 7 degrees:", seismic_response_7)
print("Seismic Response for Theta = 25 degrees:", seismic_response_25)
```

```python
def calculate_vsh(**kwargs):

    GR_log = kwargs.get("GR_log", 0)
    GR_clean = kwargs.get("GR_clean" , 0)
    GR_shale = kwargs.get("GR_shale", 1)

    Vsh = (GR_log - GR_clean) / (GR_shale - GR_clean)
    return round(Vsh, 4)
```

```python
def calc_SW(Rt, *porosity_values, a=1, m=2, n=2, **kwargs):

    Rw = kwargs.get("Rw", 0.1)

    print(f"porosity values : {porosity_values}")
    print(f"Archie constants: a =  {a}, m = {m}, n = {n}")
    print(f"Additional information: {kwargs}")

    for phi in porosity_values:
        Sw = ((a *Rw) / (phi **m *Rt))**(1/n)
        print(f"\nfor porosity value {phi:.2f}: water saturation is {Sw:.3f}\n")
```

```python
rock_densities = [1.8, 2.5, 3.2, 2.8, 1.9, 3.5]
rock_types = []

def rock_classification(density):
    for value in density:
        if value < 2:
            rock_types.append((value,"Sedimentary"))
        elif 2 < value <= 3:
            rock_types.append((value, "Metamorphic"))
        elif value > 3:
            rock_types.append((value, "Igneous"))

    print(rock_types)
```

```python
rock_classification(rock_densities)
```

```python
rock_densities = [1.8, 2.5, 3.2, 2.8, 1.9, 3.5]

rock_types = [(d, "Sedimentary") if d < 2.0 else (d, "Metamorphic") if d <= 3.0 else (d, "Igneous") for d in rock_densities]

#[expression for item in iterable if condition ]

#[[if condition else condition else condition expression for item in iterable if condition ]]
print(rock_types)
```

**Filter Function**

```python
#filter(function, iterable)

seismic_amplitudes = [30, 55, 42, 67, 90, 25, 51, 75]
high_amplitudes = []

def is_high_amplitude(values):
    for amp in values:
        if amp > 50:
            high_amplitudes.append(amp)
    print(high_amplitudes)

is_high_amplitude(seismic_amplitudes)
```

```python
seismic_amplitudes = [30, 55, 42, 67, 90, 25, 51, 75]

high_amplitudes = [amp for amp in seismic_amplitudes if amp > 50]

print(high_amplitudes)
```

```python
seismic_amplitudes = [30, 55, 42, 67, 90, 25, 51, 75]

high_amplitudes = lambda amplitudes : [x for x in amplitudes if x > 50]

print(high_amplitudes(seismic_amplitudes))
```

```python
seismic_amplitudes = [30, 55, 42, 67, 90, 25, 51, 75]

def high_amp(Amplitude_list):
    return Amplitude_list > 50



high_amplitudes = list(map(high_amp, seismic_amplitudes))

print(high_amplitudes)
```
