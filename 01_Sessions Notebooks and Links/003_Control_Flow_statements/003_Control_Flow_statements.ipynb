```python
#If eilf

Grain_Size = float(input('What is the grain size in mm?'))

if Grain_Size <= 0.002:
  print('Soil Type is Clay')
elif 0.002 < Grain_Size <= 0.075:
  print('Soil Type is Silt')
elif 0.075 < Grain_Size <= 0.42:
  print('Soil Type is Fine Sand Stone')
elif 0.42 < Grain_Size <= 2.0:
  print('Soil Type is Medium Sand Stone')
elif 2.0 < Grain_Size <= 4.75:
  print('Soil Type is Coarse Sand Stone')
elif 4.75 < Grain_Size <= 75:
  print('Soil Type is Gravel')
else:
  print('Out of scale')
```

```python
Max_Pres = 5000 #psi
Reservoir_Pressure = 3000 #psi
while Reservoir_Pressure < Max_Pres:
  if Reservoir_Pressure<= 3500:
    print(f'Close to Normal pressure. Crrent pressure is {Reservoir_Pressure} psi')
  elif 4500 >= Reservoir_Pressure >3500:
    print(f'Higher than Normal Pressure. Crrent pressure is {Reservoir_Pressure} psi')
  elif Reservoir_Pressure >4500:
    print(f'Close to Maximum Pressure. Crrent pressure is {Reservoir_Pressure} psi')
  Reservoir_Pressure = Reservoir_Pressure + 100
else:
  print(f'Maximum Pressure Reached. Current pressure is {Reservoir_Pressure} psi')
```
