```python
depth = 0
total_Depth = 3000
porosity = float(input('What is the porosity value at surfec depth? '))
Decay_Factor = 0.00009


while depth <= total_Depth:
    porosity = porosity* (1-Decay_Factor*depth)
    depth += 500
print(f"Depth : {depth} | porosity: {porosity:.4f}")
```

```python
total_Depth = 3000
Decay_Factor = 0.00009

for depth in range(0, total_Depth+1, 500):
    porosity = float(input('What is the porosity value at surfec depth? '))
    porosity = porosity* (1-Decay_Factor*depth)
    print(f"Depth : {depth} | porosity: {porosity:.4f}")
```

```python
def OGIP():
    GBV = float(input("GBV: "))
    N_G = float(input("Net to Gross: "))
    por = float(input("Porosity"))
    Sw = float(input("Water Saturation:"))
    FVF = float(input("FVF: "))
    OGIP = ((GBV*N_G*por*(1-Sw))/FVF)

    return f'Original Gas in Place is {OGIP}'

```
